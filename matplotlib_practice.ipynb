```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
```

```python
# set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# set the y-axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
#create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label = 'Boston')

#create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# set the limit between 0 and 45.
plt.ylim(0, 45)

#create a title
plt.title("PyBer Fare by Month")

#add the legend
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#create the plt
plt.barh(x_axis, y_axis, color ='magenta', label = 'Boston')
plt.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
plt.gca().invert_yaxis()
```

```python
plt.scatter(x_axis, y_axis, color = "red", label = "Chicago", s= [i * 3 for i in y_axis])
plt.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.gca().invert_yaxis()



plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s= [i * 5 for i in y_axis], label='Boston')

plt.legend()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.title("PyBer Fare by Month")
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
#assign 12 colors, one for each month
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose" ]
```

```python
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, 
        explode=explode_values, 
        labels=x_axis, 
        colors = colors,
        autopct='%.1f%%')
plt.show()
```
